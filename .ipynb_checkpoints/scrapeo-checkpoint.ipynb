```python
import requests
from lxml import html
```

```python
web=requests.get('https://www.recetasgratis.net/Macarrones-busqCate-1.html')
tree=html.fromstring(web.content)
```

```python
tree
```

```python
urls=tree.xpath('//div[@class="resultado link"]/a/@href')
```

```python
urls
```

```python
textos=[]
for a in urls:
    web=requests.get(a)
    tree=html.fromstring(web.content)
    info=tree.xpath('//li[@class="ingrediente"]/label/text()')
    for b in info:
        textos.append(b.strip())
```

```python
textos
```
