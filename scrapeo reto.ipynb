```python
frases=[]
for a in textos:
    for b in a:
        if len(b.strip()) > 0:
            print(len(b),b)
            frases.append(b)
    
```

```python
frases
```

```python
frases_limpio=[]
for a in frases:
    frases_limpio.append(a.replace(",",""))
    frases_limpio.append(a.replace('"',''))
    frases_limpio.append(a.replace('(',''))
    frases_limpio.append(a.replace(')',''))
```

```python
palabras=[]
for a in frases_limpio:
    print(a.split())
    palabras.append(a.split())
```

```python
palabras
```

```python
palabras_bien=[]
for a in palabras:
    for b in a:
        palabras_bien.append(b)
```

```python
palabras_bien
```
